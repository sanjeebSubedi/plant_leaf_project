```python
class_names = {'Anhui_Barberry': 0, "Beale's_barberry": 1, 'Big-fruited_Holly': 2, 'Canadian_poplar': 3, 'Chinese_Toon': 4, 'Chinese_cinnamon': 5, 'Chinese_horse_chestnut': 6, 'Chinese_redbud': 7, 'Chinese_tulip_tree': 8, 'Crape_myrtle,_Crepe_myrtle': 9, 'Ford_Woodlotus': 10, 'Glossy_Privet': 11, 'Japan_Arrowwood': 12, 'Japanese_Flowering_Cherry': 13, 'Japanese_cheesewood': 14, 'Japanese_maple': 15, 'Nanmu': 16, 'camphortree': 17, 'castor_aralia': 18, 'deodar': 19, 'ginkgo,_maidenhair_tree': 20, 'goldenrain_tree': 21, 'oleander': 22, 'peach': 23, 'pubescent_bamboo': 24, 'southern_magnolia': 25, 'sweet_osmanthus': 26, 'tangerine': 27, 'trident_maple': 28, 'true_indigo': 29, 'wintersweet': 30, 'yew_plum_pine': 31}
```

```python
def _get_key(val):
    for key, value in class_names.items():
        if val == value:
            return key
    return None
```

```python
image_dir = r"C:\\Users\\Windows\\Desktop\\college_project\\test\\castor_aralia\\1387_90.jpg"
```

```python
image = cv2.imread(image_dir)
```

```python
image_normalized = image/255.0
```

```python
image_sized = cv2.resize(image_normalized,(224,224))
```

```python
image_reshaped = np.reshape(image_sized,(1,224,224,3))
```

```python
output = model.predict(image_reshaped)
```

```python
plt.imshow(image_sized)
print(_get_key(np.argmax(output)))
```
